## Ejercicio 1:  Una variable numérica.
* Creo un DataFrame utilizando una consulta SQL que me muestre las toda la información de las transacciones que sean superiores a 450 y que sean de Alemania para lo que utilizaré las tablas transactions y companies. 

```python
consulta1 = "SELECT * FROM transactions JOIN companies ON business_id = company_id WHERE amount > 450 AND country = 'Germany' "

df = pd.read_sql(consulta1, conexion)
df.info()
df.head(10)
```

* GRÁFICA DE LÍNEAS: Para representar esta gráfica tomaré la columna amount del DataFrame que se ha generado en el paso anterior.

```python
plt.plot(df['amount'], label='Transactions', color='green',marker='*', linestyle='--')
plt.legend()
plt.title('Transacciones superiores a 450 efectuadas en Alemania')
plt.xlabel('Conteo de ransacciones')
plt.ylabel('Importe')
plt.yticks(np.arange(450, 510, 10))  # ticks son los saltos (np.arange(desde,hasta,saltos))
plt.grid()  
plt.show()
```

## Ejercicio 2: Dos variables numéricas.  
* Hago la consulta SQL tomando los campos user_id y amount de la tabla transactions que almacenaré en la variable consulta2 para despues convertirla en un DataFrame.

```python
consulta2 = 'SELECT user_id, amount FROM transactions '
df = pd.read_sql(consulta2, conexion)
df.info()
df.head(10)
```

* GRÁFICO DE DISPERSIÓN: Para este gráfico tomare las 2 columnas que obtuve de la consulta SQL anterior y pondré en el eje X los id de usuario y en el eje Y los importes de las transacciones.

```python
plt.scatter(df['user_id'],df['amount'])
plt.xlabel('ID de Usuario')
plt.ylabel('Importe')
plt.title('Compras hechas por usuarios')
plt.show()
```

## Ejercicio 3: Una variable categórica.  
* En este caso haré una consulta SQL que me muestre la cantidad de usuarios que tiene cada país para con el resultado poder crear un DaraFrame al que le pondré de índice los nombres de los países.

```python
consulta3 = 'SELECT country, count(id) As conteo FROM users GROUP BY country'
df = pd.read_sql(consulta3,conexion)
df = df.set_index('country')
df.info()
df
```

* GRÁFICO DE PASTEL: 

```python
conteo = df['conteo']
country = df.index  
plt.figure(figsize=(6, 6))
plt.pie(conteo, labels=country, autopct='%1.1f%%', startangle=190)
plt.title('Usuarios por País')
plt.legend(title = "Países", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.axis('equal')
plt.show()
```
