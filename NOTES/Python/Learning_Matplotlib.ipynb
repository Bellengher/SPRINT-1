# Creo una lista de aos donde evaluar la puntuacin obtenida por los estudiantes.  
* Las notas de los estudiantes las genero aleatoreamente con numpy.

```python
import numpy as np

aos =['2011','2012','2013','2014','2015','2016','2017','2018','2019','2020']

nivel1 = np.random.rand(10)*100
nivel2 = np.random.rand(10)*200+100
nivel3 = np.random.rand(10)*300+200
nivel4 = np.random.rand(10)*400+300
nivel5 = np.random.rand(10)*500+400

# Ya que tengo los datos paso a graficarlos.
```

* Pasamos a hacer las grficas:
# Grficas en lneas (plot):

```python
# Introducir series de datos con color, marcador, etiqueta y estilo:

plt.plot(aos, nivel1, label='Nivel 1', color='purple',marker='<', linestyle='-')
plt.plot(aos, nivel2, label='Nivel 2', color='red',marker='*', linestyle='--')
plt.plot(aos, nivel3, label='Nivel 3', color='blue',marker='^', linestyle=':')
plt.plot(aos, nivel4, label='Nivel 4',color='green',marker='.', linestyle='-.')
plt.plot(aos, nivel5, label='Nivel 5',color='black',marker='+', linestyle=' ')

# Activar leyendas
plt.legend()

# Ttulo de grfico
plt.title('Examen de certificacin')

# Etiqueta de los ejes
plt.xlabel('Aos que se ha realizado el exmen')
plt.ylabel('Puntaje')

# Personalizar Eje de Puntaje de 0 a 1000 en incrementos de 50
plt.yticks(np.arange(0, 1001, 50))  # ticks son los saltos (np.arange(desde,hasta,saltos))

# Activar cuadrcula
plt.grid()


# Activar marcas menores
plt.minorticks_on

plt.show()
```

# Grficas de barras verticales (bar):

```python
eje_x = np.arange(0,10)

plt.bar(eje_x, nivel5, width=1/5)
plt.bar(eje_x+0.2, nivel4, width=1/5)   # +0.2 es para desfazar las barras , with=1/5 divide el ancho entre 5 para que no se solapen
plt.bar(eje_x+0.4, nivel3, width=1/5)
plt.bar(eje_x+0.6, nivel2, width=1/5)
plt.bar(eje_x+0.8, nivel1, width=1/5)

plt.show()
```

# Grfica de Barras Horizontales (barh):

```python
plt.barh(eje_x+0.8, nivel5, height= 1/5)
plt.barh(eje_x+0.6, nivel4, height= 1/5)
plt.barh(eje_x+0.4, nivel3, height= 1/5)
plt.barh(eje_x+0.2, nivel2, height= 1/5)
plt.barh(eje_x,     nivel1, height= 1/5)

plt.show()
```

# Grfica de Barras Verticales Apiladas:

```python
plt.bar(eje_x, nivel5)
plt.bar(eje_x, nivel4, bottom=nivel5)
plt.bar(eje_x, nivel3, bottom=nivel4+nivel5)
plt.bar(eje_x, nivel2, bottom=nivel3+nivel4+nivel5)
plt.bar(eje_x, nivel1, bottom=nivel2+nivel3+nivel4+nivel5)

plt.show()
```

# Grfica de Dispersin (scatter):

```python
plt.scatter(eje_x, nivel5)
plt.scatter(eje_x, nivel4)
plt.scatter(eje_x, nivel3)
plt.scatter(eje_x, nivel2)
plt.scatter(eje_x, nivel1)

plt.show()
```

# Grfica de Pie (pie):

```python
plt.pie(nivel1,
        labels=['Mxico','Colombia','Espaa','Estados Unidos','Ecuador','Argentina','Venezuela','Per','Honduras','Uruguay'])

plt.show()
```

```python
plt.bar(eje_x,     nivel1, width= 1/5)
plt.bar(eje_x+0.2, nivel2, width= 1/5)

plt.plot(aos, nivel3)

plt.scatter(eje_x, nivel4)
plt.scatter(eje_x, nivel5)

plt.show()
```
